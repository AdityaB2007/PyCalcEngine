# PyCalcEngine: A Numerical Calculus Engine in Python

PyCalcEngine is a small Python project built with **NumPy** and **Matplotlib** that allows you to:

- Compute **numerical derivatives** using finite differences  
- Compute **numerical integrals** using the trapezoidal rule  
- Visualize functions, derivatives, and cumulative integrals  
- Experiment with **polynomials, trigonometric functions, exponential functions**, and more  

This notebook demonstrates **numerical methods and applications of Python** in action.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def derivative(f, x, h=1e-10):
    """
    Approximate derivative of f at points x using finite differences.
    """
    return (f(x + h) - f(x)) / h
```

```python
def integral(f, a, b, n=1000):
    """
    Approximate integral of f from a to b using trapezoidal rule.
    """
    x = np.linspace(a, b, n)
    y = f(x)
    dx = (b - a)/n
    return np.sum((y[:-1] + y[1:])/2 * dx)
```

```python
def cumulative_integral(f, a, b, n=1000):
    """
    Compute cumulative integral for plotting.
    Returns x points and cumulative integral values.
    """
    x = np.linspace(a, b, n)
    y = f(x)
    dx = (b - a)/n
    cum_int = np.cumsum((y[:-1] + y[1:])/2 * dx)
    return x[:-1], cum_int
```

```python
def plot_function(f, a, b, n=1000):
    """
    Plot function, derivative, and cumulative integral.
    """
    x = np.linspace(a, b, n)
    y = f(x)
    dy_dx = derivative(f, x)
    x_int, cum_int = cumulative_integral(f, a, b, n)

    plt.figure(figsize=(10,6))
    plt.plot(x, y, label="f(x)")
    plt.plot(x, dy_dx, label="f'(x)")
    plt.plot(x_int, cum_int, label="Integral f(x) dx")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Function, Derivative, and Integral")
    plt.legend()
    plt.grid(True)
    plt.show()
```

## Examples

Below are a few example functions demonstrating **derivative, integral, and visualization**.

```python
# Polynomial example
f_poly = lambda x: x**3 + 2*x**2 - x
print("Integral of f_poly from 0 to 2:", integral(f_poly, 0, 2))
plot_function(f_poly, -5, 5)
```

```python
# Trigonometric example
f_sin = np.sin
print("Integral of sin(x) from 0 to pi:", integral(f_sin, 0, np.pi))
plot_function(f_sin, 0, 2*np.pi)
```

```python
# Exponential example
f_exp = np.exp
print("Integral of exp(x) from 0 to 1:", integral(f_exp, 0, 1))
plot_function(f_exp, 0, 3)
```
